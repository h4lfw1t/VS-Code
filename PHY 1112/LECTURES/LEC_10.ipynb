# Using matplotlib

## Introduction

Matplotlib is a simple library for creating plots and graphs. It is a very powerful library and is used by many data scientists and researchers. It is also used by many other libraries such as pandas and seaborn.

One of the key parts of making a graph with matploblib is using some sort of data storing method. In class, we talk about how you can store this data in an array, or as shown in class for the graph of cos(x), using np.linspace and specifiying the range of the x values. Recall that linspace is a function that creates an array of evenly spaced numbers over a specified range, including the endpoints. Having arranged our x values, we can then create a corresponding array for our y values using the function we want to graph (np.cos(x) in this case).
The full plot can be generated using the plt.plot() function, which takes in the x and y values as arguments. The plt.show() function is then used to display the plot.

To add labels to the x and y axes, we can use the plt.xlabel() and plt.ylabel() functions. We can also add a title to the graph using the plt.title() function. Changing the scale of the x and y axes can be done using the plt.xlim() and plt.ylim() functions. We can also add a grid to the graph using the plt.grid() function. Titles can be added with the plt.title() function, and legends can be added using the plt.legend() function.

```python
import matplotlib.pyplot as plt
import numpy as np

# Specify the x-axis values
x = np.linspace(0, 2*np.pi)

# Create the figure. A figure is the window in which the graph is drawn, and it can contain multiple graphs.
plt.figure(1)

# Create the graph. The first argument is the x-axis values, and the second argument is the y-axis values.
plt.plot(x, np.cos(x))

# Draw the x-axis and y-axis at y=0 and x=0, respectively. The color is red, the line style is dashed, and the line width is 2.
plt.hlines(0, 0, 2*np.pi, colors='r', linestyles='dashed', linewidths=2)

# Label the graph
plt.title('y = cos(x)')
plt.xlabel('x (radians)')
plt.ylabel('y')

# Set the x-axis limits
plt.xlim(0, 2*np.pi)

# Set the x-axis tick values and labels
xtick_values = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
xtick_labels = ['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
plt.xticks(xtick_values, xtick_labels)

# Add a grid
plt.grid()

plt.show()
```
