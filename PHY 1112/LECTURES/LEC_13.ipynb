# Vectors and Vector Fields
Vectors are a fundamental concept in physics and engineering. They are used to represent physical quantities that have both magnitude and direction. For example, force, velocity, and acceleration are all vector quantities. In numpy, vectors are easy to represent using n-dimensional arrays. To represent a vector in a plot, we can use many different methods, some easier than others.
A more complicated representation of vectors is the vector field. A vector field is a function that assigns a vector to each point in space. This is a very useful concept in physics and engineering, as it can be used to represent things like the electric field, the velocity field of a fluid, or the force field of a gravitational field. In this notebook, we will explore how to represent vectors and vector fields in python using numpy and matplotlib.
The following is an example of how to represent the electrical field of a spherical charge in space.

Using the formula for the electric field of a point charge, we can represent the electric field of a point charge in space. The electric field of a point charge is given by the formula:

E(r) = 1/(4πε₀) * q/r² * r̂ r > R, 1/(4πε₀) * q/R² * r̂ r < R

where E(r) is the electric field at a point r, q is the charge of the point charge, r̂ is the unit vector pointing from the charge to the point r, and ε₀ is the vacuum permittivity.

We can represent the electric field of a point charge in space using a vector field. We can represent the electric field at each point in space using a vector, and then plot the vector field using a quiver plot. A quiver plot is a type of plot that represents vector fields. It is a plot that represents the vectors as arrows, with the direction and magnitude of the arrows representing the direction and magnitude of the vectors. In this notebook, we will explore how to represent the electric field of a point charge in space using a quiver plot.

Using meshplot, we can represent the electric field of a point charge in space. Meshplot takes a set of points in space and a set of vectors, and then plots the vectors at the points.

Following the meshplot, we can then plug in the formula for the electric field of a point charge and plot the electric field of a point charge in space using a quiver plot. A quiver plot is a type of plot that represents vector fields. It is a plot that represents the vectors as arrows, with the direction and magnitude of the arrows representing the direction and magnitude of the vectors.

```python
import numpy as np
import matplotlib.pyplot as plt

def E_field_uniform_sphere_of_charge(x_grid, y_grid, charge, radius):
    '''
    (meshgrid, meshgrid, float, float) -> (meshgrid, meshgrid)
    This function takes in two meshgrids representing the x and y coordinates of a 2D plane, a charge, and a radius. It returns the electric field at each point in the plane due to a uniformly charged sphere of radius 'radius' and charge 'charge.'
    Preconditions: x_grid and y_grid are the same shape, charge and radius are positive floats.
    '''
    norm = np.sqrt(x_grid**2 + y_grid**2)       # distance from the origin to each point on the plane.
    r3 = np.where(norm < radius, radius**3, norm**3) # if the point is inside the sphere, set r3 to radius^3, else set it to norm^3 to avoid division by zero errors.
    return charge * x_grid / r3, charge * y_grid / r3   # return the x and y components of the electric field at each point.

# define the range of the x and y coordinates of the plane.
x = np.linspace(-5,5, num=20)
y = np.linspace(-5,5, num=20)

# create a meshgrid of the x and y coordinates such that every point in the plane is represented by a pair of x and y coordinates.
x_grid, y_grid = np.meshgrid(x, y)

# define an arbitrary radius for example purposes
radius = 2

# define the field at each point in the plane due to the sphere of charge.
Ex, Ey = E_field_uniform_sphere_of_charge(x_grid, y_grid, 1, radius)

# initialize the figure
plt.figure(1)
# plot the electric field vectors
plt.quiver(x_grid, y_grid, Ex, Ey)
# plot the sphere of charge for reference
plt.gca().add_patch(plt.Circle((0,0), radius, fill=False))
# set the limits of the plot
plt.xlim(-5,5)
plt.ylim(-5,5)
# set the aspect ratio to be equal
plt.gca().set_aspect('equal', adjustable='box')
# set the title
plt.title('Electric field due to a uniformly charged sphere')
# set the x and y labels
plt.xlabel('x')
plt.ylabel('y')
# show the plot
plt.show()
```

We can represent the electric field with a contour plot. A contour plot is a type of plot that represents the electric field as a set of contours. The contours represent the magnitude of the electric field at each point in space. The contours are lines that connect points of equal electric field magnitude. In this notebook, we will explore how to represent the electric field of a point charge in space using a contour plot.

```python
import numpy as np
import matplotlib.pyplot as plt

def E_field_uniform_sphere_of_charge(x_grid, y_grid, charge, radius):
    '''
    (meshgrid, meshgrid, float, float) -> (meshgrid, meshgrid)
    This function takes in two meshgrids representing the x and y coordinates of a 2D plane, a charge, and a radius. It returns the electric field at each point in the plane due to a uniformly charged sphere of radius 'radius' and charge 'charge.'
    Preconditions: x_grid and y_grid are the same shape, charge and radius are positive floats.
    '''
    norm = np.sqrt(x_grid**2 + y_grid**2)       # distance from the origin to each point on the plane.
    r3 = np.where(norm < radius, radius**3, norm**3) # if the point is inside the sphere, set r3 to radius^3, else set it to norm^3 to avoid division by zero errors.
    return charge * x_grid / r3, charge * y_grid / r3   # return the x and y components of the electric field at each point.

# define the range of the x and y coordinates of the plane.
x = np.linspace(-5,5, num=100)
y = np.linspace(-5,5, num=100)

# create a meshgrid of the x and y coordinates such that every point in the plane is represented by a pair of x and y coordinates.
x_grid, y_grid = np.meshgrid(x, y)

# define an arbitrary radius for example purposes
radius = 2

# define the field at each point in the plane due to the sphere of charge.
Ex, Ey = E_field_uniform_sphere_of_charge(x_grid, y_grid, 1, radius)

# calculate the magnitude of the electric field at each point
E_mag = np.sqrt(Ex**2 + Ey**2)

# generate the contour plot
plt.contourf(x_grid, y_grid, E_mag, levels=20, cmap='jet')  # levels=20 means that there will be 20 contour lines, cmap='jet' is the color map.
plt.colorbar(label='Electric Field Magnitude')
plt.gca().set_aspect('equal', adjustable='box')   # set the aspect ratio to be equal, we can use plt.axis('equal') as well.
plt.title('Contour Plot of Electric Field')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
