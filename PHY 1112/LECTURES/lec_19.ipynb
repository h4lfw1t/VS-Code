## 2-variable Euler's method
$$
\frac{du}{dx} = {f(x, u, v)}, \frac{dv}{dx} = {g(x, u, v)}

$$
$$

{y(x)} = {u(x)} + {v(x)}
\frac{dy}{dx} = \frac{du}{dx} + \frac{dv}{dx}

$$
$$

\frac{du}{dx} = {u(x_{i+1}) = u(x_i) + h*f(x_i, u(x_i), v(x_i))}

$$
$$

\frac{dv}{dx} = {v(x_{i+1}) = v(x_i) + h*g(x_i, u(x_i), v(x_i))}
$$

## Second order ODE
$$
\frac{d^2y}{dx^2} = f(x, y, \frac{dy}{dx})
$$
$$
v(x) = \frac{dy}{dx}(x)
$$
$$
u(x) = y(x)
$$
$$
\frac{du}{dx} = \frac{dy}{dx} = v(x)
$$
$$
\frac{dv}{dx} = \frac{d^2y}{dx^2} = f(x, u(x), v(x))
$$
$$
y(x_{i+1}) = u(x_{i+1})
$$
$$
\frac{dy}{dx}(x_{i+1}) = v(x_{i+1})
$$
$$
u(x_{i+1}) = u(x_i) + h*v(x_i)
$$
$$
v(x_{i+1}) = v(x_i) + h*f(x_i, u(x_i), v(x_i))
$$



## Solve the following ODE using Euler's method
$$\frac{d^2x}{dt^2} = \frac{1}{x}(\frac{dx}{dt})^2 + 2(\frac{dx}{dt}) - x^3e^{-4t}$$

Let $$ u(t) = x(t) $$ and $$ v(t) = \frac{dx}{dt}(t) $$

Then we have the following system of ODEs
$$ \frac{du}{dt} = \frac{dx}{dt} = v(t) $$
$$ \frac{dv}{dt} = \frac{d^2x}{dt^2} = \frac{1}{u(t)}v^2(t) + 2v(t) - u^3(t)e^{-4t} $$
$$ x(t_{i+1}) = u(t_{i+1}) $$
$$ \frac{dx}{dt}(t_{i+1}) = v(t_{i+1}) $$
$$ u(t_{i+1}) = u(t_i) + h*v(t_i) $$
$$ v(t_{i+1}) = v(t_i) + h*(\frac{1}{u(t_i)}v^2(t_i) + 2v(t_i) - u^3(t_i)e^{-4t_i}) $$
